```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
```

$$\int_{1}^2 e^{-x^2/2}dx$$

```python
#constants
darts = 100000
num_in = 0

#arrays
red_dots_x = np.array([])
red_dots_y=np.array([])
blue_dots_x = np.array([])
blue_dots_y=np.array([])

def func(x):
    return math.exp(-x*x/2)

for dart in range(darts):
  x = random.uniform(1,2)
  y = random.uniform(0,0.61)
  if y < func(x):
      red_dots_x=np.append(red_dots_x, x)
      red_dots_y=np.append(red_dots_y, y)
  else:
      blue_dots_x=np.append(blue_dots_x, x)
      blue_dots_y=np.append(blue_dots_y, y)

#length*height*probability
print(func(1)*red_dots_x.size/darts)

fig=plt.figure()
ax=fig.add_subplot()
ax.set_aspect(1)
plt.scatter(red_dots_x,red_dots_y, s=0.1,c="red")
plt.scatter(blue_dots_x,blue_dots_y, s=0.1,c="blue")
plt.show()
```
